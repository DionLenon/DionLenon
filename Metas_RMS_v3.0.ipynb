### Logando no sistema do RMS 

```python
#2 Classe de acesso ao sistema do RMS clicando no icone de atalho na área de trabalho localizado a esquerda da tela
# Acessar o sistema RMS

class Logar_RMS:

    # metodo principal com 2 atributos, já imputados sem interação com o usuário
    
    def __init__(self, nome, senha):
        
        # Atributos da classe "Logar_RMS", que recebe os dois parâmetros do usuário
        
        self.nome = nome
        self.senha = senha
            
    # Metodo para acessar o RMS com um atalho na área de trabalho no canto esquerdo superior com as coordenadas plicadas abaixo
    
    def acessar_atalho(self):
                
        pyautogui.click(x=42, y=245, clicks=2)
        
    # Metodo de precionar a tecla "TAB" do teclado
    
    def tab(self):
                        
        pyautogui.press("tab")
        
    # Metodo de precionar a tecla "ENTER" do teclado
    
    def enter(self):
                
        pyautogui.press("enter", presses=1, interval=0.3)
```

```python
#3 Está função realizar a chamada da Classe Logar_RMS e passar os parâmetros do usuário para logar no sistema
# Função lojar passando dados do usuário como nome e senha

def logar():
        
    # criando três variáveis que será usadas para receber informações do usuário e todas as caracteristicas da classe Logar_RMS
    try:
        
        usuario = pyautogui.prompt("Informe o nome de Usuário", title="Dados do Usuário")
                
        if usuario != None and usuario != "":
                        
            try:
                
                senha = pyautogui.password("Informe a senha do Usuário", title="Senha do Usuário", mask="*")
                
                if senha != None and senha !="":
                    
                    acessar = Logar_RMS(usuario,senha)
                    
                    # Executando o metodo da classe

                    acessar.acessar_atalho()

                    # Aplicando um pausa de 2 segundos antes de executar o próximo código

                    time.sleep(2)

                    # Digitando o nome e a senha do usuário imputados na varíavel "acessar", que recebe os atributos das classe "Logar_RMS"

                    pyautogui.write (str("{}\t".format(acessar.nome)),interval=0.2)
                    pyautogui.write (str("{}\t".format(acessar.senha)),interval=0.2)

                    # Executando o metodo da classe

                    acessar.tab()

                    # Executando o metodo da classe

                    acessar.enter()

                else:
                    sys.exit()
            
            except SystemExit:
                
                pyautogui.alert(text="Senha não informada!\nSistema Encerrado!", title="Fechamento do Sistema!")
        
        else:
            sys.exit()   
            
    except SystemExit:
                
        pyautogui.alert(text="Usuário não informado!\nSistema Encerrado!", title="Fechamento do Sistema!")
```

### Digitando as participações das lojas

```python
#5 Uma pausa de meio segundo a cada linha de código

def dig_meta_part():

    pyautogui.PAUSE = 0.5

    # Aguarda 2 segundos para iniciar o próximo código
    
    time.sleep(2)
    
    # Criando uma variavel e colocando os dados da planilha nela
    
    planilha = pd.read_excel('meta_lojas.xlsx', sheet_name = 'lojas')

    # Criando uma variavel e garantindo o inicio correto do loop
    
    index = 0

    # Guarda os valores da primeira linha de cada coluna
    
    var_loja = planilha.loc[index,"lj"]
    var_loja = int (var_loja)
    meta_loja = planilha.loc[index,"part vda"]
    mar_loja = planilha.loc[index,"mg"]

    # Digitando o F2 para alterar os valores da loja
    
    pyautogui.press("f2")

    # Digitando o número loja na primeira sequência
    
    pyautogui.write (str("{}".format(var_loja)), 0.2)# Garantindo a estrutura da variavel

    # digita Enter
    
    pyautogui.press("enter")

    # Digita a Meta da loja
    
    pyautogui.write (str("{:.6f}".format(meta_loja)), 0.2)

    # Digita 3 enter para chegar até a margem da loja
    
    pyautogui.press("enter", presses=3, interval=0.3)

    # Digita a Margem da loja
    
    pyautogui.write (str("{:.2f}".format(mar_loja)), 0.2)
    pyautogui.press("enter")

    # Confirma as alterações da participação e da margem da loja
    
    pyautogui.press("f4")
    pyautogui.press("enter")

    # Aguarda 1 segundos para iniciar o próximo código
    
    time.sleep(1)

    # Criando uma função para ser usada no loop
    
    def digitar_meta ():
        
        # Alterando os valores das variaveis com os novos dados da nova posição do index dentro do loop
        
        var_loja = planilha.loc[index,"lj"]
        var_loja = int (var_loja)
        meta_loja = planilha.loc[index,"part vda"]
        mar_loja = planilha.loc[index,"mg"]

        # Digita o número da loja
        
        pyautogui.write (str("{}".format(var_loja)), 0.2)
        pyautogui.press("enter")
        pyautogui.press("f2")

        # Digita o valor de participação da loja
        
        pyautogui.write (str("{:.6f}".format(meta_loja)), 0.2)
        pyautogui.press("enter", presses=3, interval=0.3)

        # Digita o valor da Margem da loja
        
        pyautogui.write (str("{:.2f}".format(mar_loja)), 0.2)
        pyautogui.press("enter")
        pyautogui.press("f4")
        pyautogui.press("enter")

    # Entrando no loop para digitar as demais metas
    # Obs: A planilha possui no final um linha com o total de algumas colunas o qual precisamos contar o max -1 para não pegar essa informação
    
    while index <= max(planilha.index)-1:
        
        # verifica se o valor da variável (var_loja) é difente do valor da posição do index, caso seja igual termina o loop 
        
        if var_loja != planilha.loc[index,"lj"]:
            digitar_meta()
        index = index+1

    # Após o término da digitação de todas as seções, volta para tela inicial
    
    pyautogui.press("f3")
```

### Digitando as metas das seções

```python
#6 Uma pausa de 0,5s segundo a cada linha de código

def dig_meta_secoes():
    
    pyautogui.PAUSE = 0.5

    # Aguarda 4 segundos para iniciar o próximo código
    
    time.sleep(2)
    
    # Clica no campo "Filial" para garantir a posição correta da inialização da digitação
    
    pyautogui.click(x=664, y=400)

    # Criando uma variavel e colocando os dados da planilha nela
    
    planilha = pd.read_excel('meta_lojas.xlsx', sheet_name= 'metas')

    # Criando uma variavel e garantindo o inicio correto do loop
    
    index = 0

    # guarda os valores da primeira linha de cada coluna
    
    varloja = planilha.loc[index,"lj"]
    varsecao = planilha.loc[index,"cod"]
    varpart = planilha.loc[index,"part vda"]
    varmg = planilha.loc[index,"mg"]

    # Digitando o número loja na primeira sequência
    #pyautogui.write (str("{}".format(varloja)), 0.2)# Garantindo a estrutura da variavel

    # digita Enter
    #pyautogui.press("enter")

    # digita F7
    #pyautogui.press("f7")

    # Digita a seção
    
    pyautogui.write (str("{}".format(varsecao)), 0.2)
    pyautogui.press("enter")

    # Digitar F2 para Alterar os valores de participação da seção
    
    pyautogui.press("f2")
    pyautogui.write (str("{:.6f}".format(varpart)), 0.2)
    pyautogui.press("enter", presses=3, interval=0.3)
    pyautogui.write (str("{:.2f}".format(varmg)), 0.2)
    pyautogui.press("enter")
    pyautogui.press("f4")
    pyautogui.press("enter")

    # Aguarda 2 segundos para iniciar o próximo código
    
    time.sleep(2)

    # Criar uma função para ser chamada dentro do loop para troca de loja
    
    def trocar_loja():
        
        # 2 vezes F3
        
        pyautogui.press("f3", presses=2, interval=0.3)

        # 5 vezes o Tab
        
        pyautogui.press("tab", presses=5, interval=0.3)

        # Digita a nova loja
        
        pyautogui.write (str("{}".format(varloja)), 0.2)
        pyautogui.press("enter")
        pyautogui.press("f7")

    # Criando outra função para digitar os % das metas das seções
    
    def digitar_secao():
        
        # Digita o número da seção
        
        pyautogui.write (str("{}".format(varsecao)), 0.2)
        pyautogui.press("enter")
        pyautogui.press("f2")
        varpart = planilha.loc[index,"part vda"]

        # Digita o Percentual de participação da seção
        
        pyautogui.write (str("{:.6f}".format(varpart)), 0.2)
        pyautogui.press("enter", presses=3, interval=0.3)

        # Digita a Margem
        
        varmg = planilha.loc[index,"mg"]
        pyautogui.write (str("{:.2f}".format(varmg)), 0.2)
        pyautogui.press("enter")
        pyautogui.press("f4")
        pyautogui.press("enter")

    # como a planilha já está na ordem não me preocupei em filtrar loja por loja e secao por secao

    while index <= max(planilha.index):
        
    # verifica se a loja mudou toda vez que muda de linha. Se mudou eu atribuo o novo valor na mesma variavel e printo
    
        if varloja != planilha.loc[index,"lj"]:
            varloja = planilha.loc[index,"lj"]
            trocar_loja()        

    # verifica se a secao mudou toda vez que muda de linhaSe mudou eu atribuo o novo valor na mesma variavel e printo
    
        if varsecao != planilha.loc[index,"cod"]:
            varsecao = planilha.loc[index,"cod"]
            digitar_secao()
        index = index+1

    # Após o término da digitação de todas as seções, volta para tela inicial
    
    pyautogui.press("f3", presses=2, interval=0.3)
```

### Digitando as metas da diariazação

```python
#7 Digitação da diariazação

def dig_diarizacao():
    
    # Aplicando uma pausa em cada linha de código e aguardando 2 segundos para iniciar as demais linhas de código
    
    pyautogui.PAUSE = 0.5
    time.sleep(2)
    
    # Clica no campo "Filial" para garantir a posição correta da inialização da digitação
    
    pyautogui.click(x=664, y=400)
    
    # Cria uma lista para o dados da semana e digitar os dados de lá
    
    planilha = pd.read_excel('meta_lojas.xlsx', sheet_name= 'metas')
    
    # Criar e carregar valores em algumas variaveis
    # Antes de iniciar a digitação
    # Variável index servirá como ponteiro para acessar cada linha da planilha de meta
    
    index = 0
    
    # Variaveis que receberão os dados da meta de acordo com o ponteiro do index na planilha de meta
    
    varloja = planilha.loc[index,"lj"]
    varsecao = planilha.loc[index,"cod"]
    varpart = planilha.loc[index,"domingo"]
    varmg = planilha.loc[index,"mg"]
    
    # Criar uma função chamada de (criar_lista) para ser usada posteriormente em outro bloco do código
    
    def criar_lista():
        
        # Criação de uma lista, o qual vai receber os dados de cada dia da semana para execução de casa seção
        
        lista = []
        lista.append("{:.6f}".format(planilha.loc[index,"domingo"]))
        lista.append("{:.6f}".format(planilha.loc[index,"segunda"]))
        lista.append("{:.6f}".format(planilha.loc[index,"terca"]))
        lista.append("{:.6f}".format(planilha.loc[index,"quarta"]))
        lista.append("{:.6f}".format(planilha.loc[index,"quinta"]))
        lista.append("{:.6f}".format(planilha.loc[index,"sexta"]))
        lista.append("{:.6f}".format(planilha.loc[index,"sabado"]))
        return lista
    
    
    # Chamando a execução da função para a criar a lista com os dados populados de cada coluna da primeira linha
    
    criar_lista()
    
    # Criar função (digitar_loja) para ser usado no próximo bloco de código
    
    def digitar_loja():
        
        # Digitando o número loja na primeira sequência
        
        pyautogui.write (str("{}".format(varloja)), 0.2)# Garantindo a estrutura da variavel
        pyautogui.press("enter")
        
        # digita F7 para acessar as seções da loja
        
        pyautogui.press("f7")
        
        # Digita a seção
    # Chamando a execução da função (digitar_loja) para a primeira digitação
    
    digitar_loja()
    
    # Criar função (digitar_secao) para ser usado no próximo bloco de código
    
    
    def digitar_secao():
        pyautogui.write (str("{}".format(varsecao)), 0.2)
        pyautogui.press("enter")
        
        # Acessando o modulo de diarização
        
        pyautogui.press("f8")
        
        # Alterando os valores da diarização
        
        pyautogui.press("f2")
        pyautogui.press("enter")
        
        # Digitando os novos valores de cada dia na primeira sequencia
        
        pyautogui.press("backspace",presses=10, interval=0.1)
        
    # Chamando a execução da função (digitar_secao) para a primeira digitação
    
    digitar_secao()
    
    # Criar outra função de digitação da diarização de cada seção
    
    
    def digitar_diarização():
        
        # Criando um loop para digitar a diariazação da primeira seção
        
        a = 0
        while a <= 6:
            pyautogui.write ('{}\n'.format((criar_lista() [a])), 0.2)
            pyautogui.press("enter", presses=2, interval=0.3)
            pyautogui.press("del")
            pyautogui.write (str("{:.2f}".format(varmg)), 0.2)
            pyautogui.press("enter")
            pyautogui.press("del")
            a = a+1
        pyautogui.press("enter", presses=2, interval=0.3)
        pyautogui.press("f3")
        
    # Chamando a função (digitar_diarização) para digitar as metas diarias de cada seção
    
    digitar_diarização()
    
    # Acessando a próxima seção
    
    while index <= max(planilha.index):
        
        # Trocando de filial, (Mudando a loja)
        
        if varloja != planilha.loc[index,"lj"]:
            varloja = planilha.loc[index,"lj"]
            pyautogui.press("f3")
            pyautogui.press("enter", presses=5, interval=0.3)
            digitar_loja()
            
        # Alterando a seção da mesma loja desde o inicio da primeira sequencia  
        
        if varsecao != planilha.loc[index,"cod"]:
            varsecao = planilha.loc[index,"cod"]
            varmg = planilha.loc[index,"mg"]
            criar_lista()
            digitar_secao()
            digitar_diarização()
        index = index+1
        
    # Retorno para a tela principal e fim da digitação de todos os dados da diariazação
    
    pyautogui.press("f3")
```

```python
# 8 Código principal que realiza a chamada de todas as Classes e funções da Aplicação

try:
           
    pyautogui.PAUSE = 0.8
    
    ini_sist = pyautogui.confirm("Você está logado no sistema?", title = "Conferir sistema", buttons = ["Sim","Não"])
        
    if ini_sist == "Sim":
        
        # Clica na tela principal das metas de participações das lojas
        
        pyautogui.click(x=84, y=175)
        
        # Perguntar se vai digitar as participações das lojas ou das seções ou diariazação
        
        modo = pyautogui.confirm("Qual modalidade vamos iniciar as digitações?", title="Escolha de Modalidade",
                                buttons=["Participação","Seções","Diarização"])
        
        # Verificando a opção escolhida e iniciando a função correspondente e executanto o processo escolhido.
        
        if modo == "Participação":
            dig_meta_part()
            
        elif modo == "Seções":
            dig_meta_secoes()
            
        elif modo == "Diarização":
            dig_diarizacao()
            
        else:
            sys.exit()

    elif ini_sist == "Não":

        # Realizando a chamada das funções "logar" e "Telas_RMS", que farão o processo de login e carregar a tela de cadastro de metas
        
        logar()
        tela_rms()
        
        # Perguntar se vai digitar as participações das lojas ou das seções ou diariazação
        
        modo = pyautogui.confirm("Qual modalidade vamos iniciar as digitações?", title="Escolha de Modalidade",
                                buttons=["Participação","Seções","Diarização"])
        if modo == "Participação":
            dig_meta_part()
            
        elif modo == "Seções":
            dig_meta_secoes()
            
        elif modo == "Diarização":
            dig_diarizacao()
            
        else:
            sys.exit()

    else:
        sys.exit()

except SystemExit:
    
    pyautogui.alert("Sistema Fechado com Sucesso!", title="Fechamento do Sistema!")
```
